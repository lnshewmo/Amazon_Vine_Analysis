```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv('vine_table.csv')
df.head()
```

```python
# create the vine_df with reviews 
update_df = df.loc[df['total_votes'] >= 20]
update_df
```

```python
helpful_votes_df = update_df.loc[update_df['helpful_votes']/update_df['total_votes'] >= 0.5]
helpful_votes_df
```

```python
vine_review_df = helpful_votes_df.loc[helpful_votes_df['vine'] == 'Y']
vine_review_df
```

```python
unpaid_review_df = helpful_votes_df.loc[helpful_votes_df['vine'] == 'N']
unpaid_review_df
```

```python
vine_count = vine_review_df['review_id'].count()
unpaid_count = unpaid_review_df['review_id'].count()
print(vine_count)
print(unpaid_count)
```

```python
vine_5stars = vine_review_df['review_id'][vine_review_df['star_rating'] == 5].count()
unpaid_5stars = unpaid_review_df['review_id'][unpaid_review_df['star_rating'] == 5].count()
print(vine_5stars)
print(unpaid_5stars)
```

```python
percentage_5star_vines = vine_5stars / vine_count * 100
percentage_5star_unpaid = unpaid_5stars / unpaid_count * 100
print(percentage_5star_vines)
print(percentage_5star_unpaid)
```

```python
vine_summary = {'Review Type': ['Vine', 'Unpaid'],
                'Total Reviews': [vine_count, unpaid_count],
                'Number 5-Star Reviews': [vine_5stars, unpaid_5stars],
                'Percentage 5-Star Reviews': [percentage_5star_vines, percentage_5star_unpaid]}
                
vine_summary_df = pd.DataFrame(vine_summary)
vine_summary_df.set_index('Review Type', inplace=True)
vine_summary_df['Percentage 5-Star Reviews'] = vine_summary_df['Percentage 5-Star Reviews'].map("{:.1f}%".format)
vine_summary_df['Total Reviews'] = vine_summary_df['Total Reviews'].map("{:,}".format)
vine_summary_df['Number 5-Star Reviews'] = vine_summary_df['Number 5-Star Reviews'].map("{:,}".format)
vine_summary_df
```
